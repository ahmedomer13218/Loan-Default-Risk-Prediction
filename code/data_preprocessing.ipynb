## Data Cleaning

### Application data

```python
# check percentage of null values
app_empty_perc = app_data.isnull().sum()/len(app_data)*100
with pd.option_context('display.max_rows', 200):
    print(app_empty_perc.sort_values(ascending = False))
```

```python
# how many columns have more than 25% missing values
more_than_25prec_empty = (app_empty_perc > 25).sum()
print(more_than_25prec_empty)
```

```python
# print the columns that have > 25% of null data
app_empty_count = app_data.isnull().sum().sort_values(ascending = False).head(more_than_25prec_empty)
print(app_empty_count.index.tolist())
```

```python
# remove columns with null > 25%
app_data_cleaned = app_data.copy()

col_name_del = (app_data_cleaned.isnull().sum()/len(app_data_cleaned))
col_name_del = list(col_name_del[col_name_del.values>=0.25].index)
# drop
app_data_cleaned.drop(labels = col_name_del,axis =1,inplace=True)  
print(app_data_cleaned.shape)
```

```python
# remove rows with null values
app_data_cleaned.dropna(inplace=True)
print(app_data_cleaned.shape)
```

```python
# save cleaned data
app_data_cleaned.to_csv(os.path.join(current_dir, data_relative_path, 'application_data_cleaned.csv'), index=False)
```
